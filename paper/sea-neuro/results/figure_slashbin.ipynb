```python
import pickle

with open("slashbin_data.pkl", "rb") as pickle_file:
    df_slashbin_data = pickle.load(pickle_file)
```

```python
df_slashbin_data
```

```python
sea = df_slashbin_data[df_slashbin_data["fs"] == "sea"]
default = df_slashbin_data[df_slashbin_data["fs"] == "default"]
```

```python
merge_cols = ["dataset", "pipeline", "n_images", "busy_writers"]
df_merged = sea.merge(default, on=merge_cols)[merge_cols + ["makespan_x", "makespan_y"]]
```

```python
from matplotlib import pyplot as plt
import numpy as np

colors = {}
colors["afni"] = "blue"
colors["spm"] = "pink"
colors["fsl"] = "yellow"


def get_color(pipeline):
    return colors[pipeline]


df_merged["color"] = df_merged["pipeline"].apply(get_color)

plt.scatter(df_merged["makespan_y"], df_merged["makespan_x"], color=df_merged["color"])
plt.plot(np.arange(25000), np.arange(25000), color="k")
plt.xlim(0, 25000)
plt.ylim(0, 25000)
```
